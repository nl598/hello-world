```python
import matplotlib.pyplot as plt
import numpy as np


def cobweb(f, x0, N, a=0, b=1):
    # plot the function being iterated
    t = np.linspace(a, b, N)
    plt.plot(t, f(t), 'k')
    
    # plot the dotted line y = x
    plt.plot(t, t, 'k:')
    
    # plot the iterates
    x, y = x0, f(x0)
    for _ in range(N):
        fy = f(y)
        plt.plot([x, y], [y, y], 'b', linewidth=1)
        plt.plot([y, y], [y, fy], 'b', linewidth=1)
        x, y = y, fy
        
    
    plt.show()
    plt.close()
    
cobweb(lambda x: 3*x-x**3, 0.01, 100)
```
